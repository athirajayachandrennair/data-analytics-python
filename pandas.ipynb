** Merge() Function ** 

```python
create two data frame df1 and df2 then it merge in one frame
```

```python
df1=pd.DataFrame({'Roll No':[1,2,3,4],'Physics':[34,56,78,39]})
df1
```

```python
df2=pd.DataFrame({'Roll No':[1,2,3,4],'Chemistry':[34,56,78,39]})
df2
```

```python
pd.merge(df1,df2, on='Roll No')
```

```python
df1=pd.DataFrame({'Roll No':[1,2,3,4],'Economics':[34,56,78,39]})
df1
```

```python
df2=pd.DataFrame({'Roll No':[1,2,3,4],'History':[34,56,78,39]})
df2
```

```python
pd.merge(df1,df2, on='Roll No')
```

```python
pd.merge(df1,df2)  # only display intersection
```

```python
df3=pd.DataFrame({'Roll No':[1,2,3,4,5],'Economics':[34,56,78,39,67]})
df4=pd.DataFrame({'Roll No':[1,2,3,6,7],'History':[34,56,78,39,49]})
```

```python
pd.merge(df3,df4)  # only display two common roll no
```

```python
pd.merge(df3,df4, how='left')   # left merge
```

```python
pd.merge(df3,df4, how='right') # right merge
```

```python
pd.merge(df3,df4, how='outer') # full join
```

** append function **

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'Roll No':[1,2,3,4,5],'Physics':[45,56,34,56,67],'Maths':[45,34,56,78,56]})
df2=pd.DataFrame({'Roll No':[6,7,8,9,10],'Physics':[47,56,34,53,62],'Maths':[40,30,56,78,52]})
print(df1)
print(df2)
```

```python
pd.concat([df1, df2])  
```

```python
result = pd.concat([df1, df2], ignore_index=True)
print(result)                                      # ignore index
```

```python
result = pd.concat([df1, df2], ignore_index=True, sort=True)
print(result)                                                  # sort column name
```
